```python
# Create genre_df and country_df first
# Extract all genres and count their frequency
genre_counts = {}
for genres in df['genres_list']:
    for genre in genres:
        if genre in genre_counts:
            genre_counts[genre] += 1
        else:
            genre_counts[genre] = 1

# Create a dataframe from the genre counts
genre_df = pd.DataFrame({
    'Genre': list(genre_counts.keys()),
    'Count': list(genre_counts.values())
}).sort_values('Count', ascending=False).reset_index(drop=True)

# Extract all countries and count their frequency
country_counts = {}
for countries in df['production_countries_list']:
    for country in countries:
        if country in country_counts:
            country_counts[country] += 1
        else:
            country_counts[country] = 1

# Create a dataframe from the country counts
country_df = pd.DataFrame({
    'Country': list(country_counts.keys()),
    'Count': list(country_counts.values())
}).sort_values('Count', ascending=False).reset_index(drop=True)

# Display the top genres and countries
print("Top 10 Genres:")
display(genre_df.head(10))
print("\nTop 5 Production Countries:")
display(country_df.head(5))

# Now continue with the feature engineering code
# 1. Financial features
# Calculate profit and ROI (Return on Investment)
df['profit'] = df['revenue'] - df['budget']
df['roi'] = df['profit'] / df['budget']
df['roi'] = df['roi'].replace([np.inf, -np.inf], np.nan).fillna(0)  # Handle division by zero

# Create log-transformed versions of skewed features
for feature in ['budget', 'revenue', 'popularity', 'vote_count']:
    df[f'{feature}_log'] = np.log1p(df[feature])  # log1p to handle zeros

# 2. Temporal features
# Extract more features from release_date
df['release_year'] = df['release_date'].dt.year
df['release_month'] = df['release_date'].dt.month
df['release_quarter'] = df['release_date'].dt.quarter
df['release_day'] = df['release_date'].dt.day
df['release_day_of_week'] = df['release_date'].dt.dayofweek
df['is_weekend_release'] = df['release_day_of_week'].isin([5, 6]).astype(int)  # 5=Sat, 6=Sun
df['is_holiday_release'] = df['release_month'].isin([11, 12, 1, 5, 6, 7]).astype(int)  # Holiday seasons

# Calculate movie age (from release to current date)
current_date = datetime.now()
df['age_in_years'] = ((current_date - df['release_date']).dt.days / 365.25)
df['age_in_years'] = df['age_in_years'].fillna(0).clip(0)  # Fix negative values and NaNs

# 3. Genre and production features
# Count number of genres, production companies, production countries
df['genre_count'] = df['genres_list'].apply(len)
df['production_company_count'] = df['production_companies_list'].apply(len)
df['production_country_count'] = df['production_countries_list'].apply(len)

# Create binary indicators for top genres
top_genres = list(genre_df.head(10)['Genre'])
for genre in top_genres:
    df[f'is_{genre.lower().replace(" ", "_")}'] = df['genres_list'].apply(lambda x: 1 if genre in x else 0)

# Create binary indicators for top production countries
top_countries = list(country_df.head(5)['Country'])
for country in top_countries:
    df[f'is_{country.lower().replace(" ", "_")}'] = df['production_countries_list'].apply(
        lambda x: 1 if country in x else 0)

# 4. Text features
# Process overview text
df['overview_length'] = df['overview'].str.len()
df['overview_word_count'] = df['overview'].str.split().str.len()
df['has_overview'] = df['overview'].str.len().gt(0).astype(int)

if 'tagline' in df.columns:
    df['tagline_length'] = df['tagline'].str.len()
    df['has_tagline'] = df['tagline'].str.len().gt(0).astype(int)

# 5. Success features (target engineering)
# Define success indicators
df['is_profitable'] = (df['revenue'] > df['budget']).astype(int)
df['high_roi'] = (df['roi'] > 1).astype(int)  # ROI greater than 100%
df['high_rating'] = (df['vote_average'] >= 7.0).astype(int)  # Rating >= 7.0

# Create a composite success score (weighted combination of metrics)
df['success_score'] = (0.4 * df['roi'].clip(0, 10) / 10 +  # Clipped and normalized ROI (40% weight)
                      0.4 * df['vote_average'] / 10 +     # Normalized rating (40% weight)
                      0.2 * df['vote_count_log'] / df['vote_count_log'].max())  # Normalized log of vote count (20% weight)

print("Feature engineering complete. New dataframe shape:", df.shape)
print("\nNew features created:")
print(list(set(df.columns) - set(['budget', 'revenue', 'runtime', 'vote_average', 'vote_count',
                                 'popularity', 'release_date', 'overview', 'tagline', 'genres',
                                 'production_companies', 'production_countries', 'original_language'])))
```

### Feature Scaling and Transformation

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 1. Identify features for scaling
numerical_features_to_scale = [
    'budget_log', 'revenue_log', 'popularity_log', 'vote_count_log',
    'vote_average', 'runtime', 'age_in_years', 'overview_length',
    'genre_count', 'production_company_count', 'production_country_count'
]

# 2. Apply standard scaling (z-score normalization)
scaler = StandardScaler()
for feature in numerical_features_to_scale:
    if feature in df.columns:
        df[f'{feature}_scaled'] = scaler.fit_transform(df[[feature]])

# 3. Apply min-max scaling (normalization to 0-1 range)
min_max_scaler = MinMaxScaler()
for feature in numerical_features_to_scale:
    if feature in df.columns:
        df[f'{feature}_norm'] = min_max_scaler.fit_transform(df[[feature]])
```

```python
# 1. One-hot encoding for original language
# First, group rare languages
language_counts = df['original_language'].value_counts()
top_languages = language_counts[language_counts > 100].index
df['original_language_grouped'] = df['original_language'].apply(
    lambda x: x if x in top_languages else 'other')

# One-hot encode the grouped languages
language_dummies = pd.get_dummies(df['original_language_grouped'], prefix='lang')
df = pd.concat([df, language_dummies], axis=1)

# 2. Binary encoding for status (if present)
if 'status' in df.columns:
    status_dummies = pd.get_dummies(df['status'], prefix='status')
    df = pd.concat([df, status_dummies], axis=1)

# Note: For genres, we already created binary indicators in the feature engineering step
```

### Text Preprocessing

```python
# Optimized version of text preprocessing for large datasets
import nltk
import re
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from tqdm import tqdm  # For progress tracking
import multiprocessing as mp  # For parallel processing

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer and stopwords once
lemmatizer = WordNetLemmatizer()
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# Optimized preprocessing function
def preprocess_text(text):
    if not isinstance(text, str) or text == '':
        return ''

    # Convert to lowercase
    text = text.lower()

    # Remove special characters and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)

    # Tokenize
    tokens = text.split()  # Simple split is faster than word_tokenize

    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

    # Join tokens back to string
    return ' '.join(tokens)

# Option 1: Process a smaller sample first for testing
sample_size = 50000  # Adjust as needed
print(f"Processing a sample of {sample_size} records first...")
sample_df = df.sample(sample_size, random_state=42)
sample_df['overview_processed'] = sample_df['overview'].apply(preprocess_text)

# Option 2: Process in batches
def process_in_batches(df, batch_size=10000):
    print(f"Processing in batches of {batch_size}...")
    processed_series = pd.Series(index=df.index, dtype='object')

    for start in tqdm(range(0, len(df), batch_size)):
        end = min(start + batch_size, len(df))
        batch = df.iloc[start:end]
        processed_series.iloc[start:end] = batch['overview'].apply(preprocess_text)

    return processed_series

# Uncomment to use batch processing
# df['overview_processed'] = process_in_batches(df)

# Option 3: Use parallel processing (can be much faster on multi-core systems)
def parallel_process(df, func, column, n_jobs=None):
    if n_jobs is None:
        n_jobs = max(1, mp.cpu_count() - 1)  # Use all cores except one

    print(f"Processing using {n_jobs} cores in parallel...")

    # Split the dataframe
    df_split = np.array_split(df[column], n_jobs)

    # Create the pool
    pool = mp.Pool(n_jobs)

    # Process in parallel
    result = pd.concat(pool.map(lambda x: x.apply(func), df_split))

    # Close the pool
    pool.close()
    pool.join()

    return result

# Uncomment to use parallel processing
# df['overview_processed'] = parallel_process(df, preprocess_text, 'overview')

# For TF-IDF, consider using a smaller number of features or a higher min_df
print("Creating TF-IDF vectorizer with optimized parameters...")
tfidf_vectorizer = TfidfVectorizer(
    max_features=1000,  # Reduced from 5000
    min_df=10,          # Increased from 5
    max_df=0.85,
    lowercase=False,    # Text is already lowercase
    token_pattern=r'\S+' # Simpler tokenization
)

# Fit on a sample if full dataset is too large
print("Fitting TF-IDF vectorizer...")
tfidf_vectorizer.fit(sample_df['overview_processed'].dropna())

print("Text preprocessing complete for sample.")
print(f"Vocabulary size: {len(tfidf_vectorizer.vocabulary_)}")

# To apply to full dataset, uncomment one of the options above
```

### Target Value Preparation

```python
# Prepare target variables for different prediction tasks
# 1. Regression target: log_revenue
df['target_revenue'] = df['revenue_log']

# 2. Binary classification target: is_successful
# Define success as having both positive ROI and above-average rating
df['target_success'] = ((df['roi'] > 0) & (df['vote_average'] > 6.5)).astype(int)

# 3. Multi-class classification target: success_level
# Fix for quantile calculation with duplicates

# For ROI - handle duplicates by using rank instead of qcut
print("ROI range before clipping:", df['roi'].min(), "to", df['roi'].max())
df['roi_clipped'] = df['roi'].clip(-1, 10)
print("ROI range after clipping:", df['roi_clipped'].min(), "to", df['roi_clipped'].max())

# Check value distribution after clipping
print("Number of entries at maximum ROI (10):", (df['roi_clipped'] >= 10).sum())
print("Number of entries at minimum ROI (-1):", (df['roi_clipped'] <= -1).sum())

# Use rank method instead of qcut for better handling of duplicate values
df['roi_rank'] = df['roi_clipped'].rank(method='first')
df['roi_quantile'] = pd.qcut(df['roi_rank'], 4, labels=False)

# For vote_average
print("\nVote average range before clipping:", df['vote_average'].min(), "to", df['vote_average'].max())
df['vote_average_clipped'] = df['vote_average'].clip(0, 10)
print("Vote average range after clipping:", df['vote_average_clipped'].min(), "to", df['vote_average_clipped'].max())

# Check value distribution after clipping
print("Number of entries at maximum vote_average (10):", (df['vote_average_clipped'] >= 10).sum())
print("Number of entries at minimum vote_average (0):", (df['vote_average_clipped'] <= 0).sum())

# Use rank method for vote_average as well
df['vote_rank'] = df['vote_average_clipped'].rank(method='first')
df['rating_quantile'] = pd.qcut(df['vote_rank'], 4, labels=False)

# Create the success level target (0-3)
df['target_success_level'] = np.floor((df['roi_quantile'] + df['rating_quantile']) / 2).astype(int)

# Print information about the target variables
print("\nTarget variables prepared:")
print("Regression target (revenue_log) distribution:")
display(df['target_revenue'].describe())

print("\nBinary classification target (is_successful) distribution:")
display(df['target_success'].value_counts(normalize=True))

print("\nMulti-class classification target (success_level) distribution:")
display(df['target_success_level'].value_counts(normalize=True))
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, f_regression, f_classif
import numpy as np

# First, check for missing values in the target variables
print("Missing values in target_revenue:", df['target_revenue'].isna().sum())
print("Missing values in target_success:", df['target_success'].isna().sum())

# 1. Prepare data for regression (predicting revenue)
regression_features = [col for col in df.columns if '_scaled' in col or col in [
    'budget_log', 'runtime', 'popularity_log', 'vote_count_log', 'vote_average',
    'genre_count', 'production_company_count', 'production_country_count',
    'release_year', 'release_month', 'release_quarter', 'is_weekend_release',
    'is_holiday_release', 'age_in_years', 'overview_length'
]]

# Filter out rows with NaN in the target for regression
valid_reg_mask = ~df['target_revenue'].isna()
X_regression = df.loc[valid_reg_mask, regression_features].fillna(0)
y_regression = df.loc[valid_reg_mask, 'target_revenue']

print(f"Using {X_regression.shape[0]} valid samples for regression feature selection")

# Check for any remaining NaN values
if X_regression.isna().any().any():
    print("Warning: There are still NaN values in X_regression after fillna()")
    # Additional handling if needed
    X_regression = X_regression.fillna(0)

# Select top 15 features for regression
try:
    selector_regression = SelectKBest(f_regression, k=min(15, X_regression.shape[1]))
    X_regression_selected = selector_regression.fit_transform(X_regression, y_regression)

    # Get selected feature names
    selected_reg_features = [regression_features[i] for i in selector_regression.get_support(indices=True)]
    print("Top features for revenue prediction:")

    # Create a dataframe with feature names and scores
    reg_feature_scores = pd.DataFrame({
        'Feature': [regression_features[i] for i in range(len(regression_features))],
        'Score': selector_regression.scores_
    }).sort_values('Score', ascending=False)

    display(reg_feature_scores.head(15))
except Exception as e:
    print(f"Error in regression feature selection: {e}")
    # Try a simpler approach with correlation
    corr_with_target = X_regression.corrwith(y_regression).abs().sort_values(ascending=False)
    print("Top features by correlation with revenue:")
    display(corr_with_target.head(15))

# 2. Prepare data for classification (predicting success)
classification_features = regression_features + [col for col in df.columns if 'is_' in col and col != 'target_success']
# Filter out columns that don't exist
classification_features = [col for col in classification_features if col in df.columns]

# Filter out rows with NaN in the target for classification
valid_class_mask = ~df['target_success'].isna()
X_classification = df.loc[valid_class_mask, classification_features].fillna(0)
y_classification = df.loc[valid_class_mask, 'target_success']

print(f"\nUsing {X_classification.shape[0]} valid samples for classification feature selection")

# Check for any remaining NaN values
if X_classification.isna().any().any():
    print("Warning: There are still NaN values in X_classification after fillna()")
    # Additional handling if needed
    X_classification = X_classification.fillna(0)

# Select top 15 features for classification
try:
    selector_classification = SelectKBest(f_classif, k=min(15, X_classification.shape[1]))
    X_classification_selected = selector_classification.fit_transform(X_classification, y_classification)

    # Get selected feature names
    selected_class_features = [classification_features[i] for i in selector_classification.get_support(indices=True)]
    print("Top features for success prediction:")

    # Create a dataframe with feature names and scores
    class_feature_scores = pd.DataFrame({
        'Feature': [classification_features[i] for i in range(len(classification_features))],
        'Score': selector_classification.scores_
    }).sort_values('Score', ascending=False)

    display(class_feature_scores.head(15))
except Exception as e:
    print(f"Error in classification feature selection: {e}")
    # For classification, point-biserial correlation can be used as an alternative
    from scipy.stats import pointbiserialr

    correlations = []
    for col in X_classification.columns:
        corr, _ = pointbiserialr(y_classification, X_classification[col])
        correlations.append((col, abs(corr)))

    class_corr_df = pd.DataFrame(correlations, columns=['Feature', 'Correlation'])
    class_corr_df = class_corr_df.sort_values('Correlation', ascending=False)
    print("Top features by correlation with success:")
    display(class_corr_df.head(15))
```

### Data Splitting and Balancing

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample

# Check if overview_processed exists
has_processed_overview = 'overview_processed' in df.columns
print(f"Does overview_processed exist in the DataFrame? {has_processed_overview}")

# 1. Split data into training, validation, and test sets
# First, create a clean dataset with only the necessary columns that exist
needed_columns = ['id'] + selected_reg_features + selected_class_features + ['target_revenue', 'target_success', 'target_success_level']
if has_processed_overview:
    needed_columns.append('overview_processed')

# Check if all necessary columns exist
missing_columns = [col for col in needed_columns if col not in df.columns]
if missing_columns:
    print(f"Warning: The following columns are missing: {missing_columns}")
    needed_columns = [col for col in needed_columns if col in df.columns]

# Create the model dataset
model_data = df[needed_columns].copy()

# Drop rows with NaN in target variables
model_data = model_data.dropna(subset=['target_revenue', 'target_success'])

# Split data: 70% train, 15% validation, 15% test
train_data, temp_data = train_test_split(model_data, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

print(f"Training set size: {train_data.shape[0]}")
print(f"Validation set size: {val_data.shape[0]}")
print(f"Test set size: {test_data.shape[0]}")

# 2. Balance the classification dataset (for the binary target)
# Check class balance
train_class_balance = train_data['target_success'].value_counts(normalize=True)
print("\nClass balance in training data:")
print(train_class_balance)

# If imbalanced (less than 40% of minority class), use resampling
if min(train_class_balance) < 0.4:
    # Separate majority and minority classes
    majority_class = train_data[train_data['target_success'] == train_class_balance.idxmax()]
    minority_class = train_data[train_data['target_success'] == train_class_balance.idxmin()]

    # Upsample minority class
    minority_upsampled = resample(minority_class,
                                 replace=True,
                                 n_samples=len(majority_class),
                                 random_state=42)

    # Combine majority and upsampled minority
    train_data_balanced = pd.concat([majority_class, minority_upsampled])

    print(f"After balancing, training set size: {train_data_balanced.shape[0]}")
    print("New class balance:")
    print(train_data_balanced['target_success'].value_counts(normalize=True))
else:
    train_data_balanced = train_data
    print("Dataset is relatively balanced, no resampling needed.")

# Save the splits to variables for later use
X_train = train_data_balanced.drop(['id', 'target_revenue', 'target_success', 'target_success_level'], axis=1, errors='ignore')
y_train_reg = train_data_balanced['target_revenue']
y_train_class = train_data_balanced['target_success']
y_train_multi = train_data_balanced['target_success_level'] if 'target_success_level' in train_data_balanced.columns else None

X_val = val_data.drop(['id', 'target_revenue', 'target_success', 'target_success_level'], axis=1, errors='ignore')
y_val_reg = val_data['target_revenue']
y_val_class = val_data['target_success']
y_val_multi = val_data['target_success_level'] if 'target_success_level' in val_data.columns else None

X_test = test_data.drop(['id', 'target_revenue', 'target_success', 'target_success_level'], axis=1, errors='ignore')
y_test_reg = test_data['target_revenue']
y_test_class = test_data['target_success']
y_test_multi = test_data['target_success_level'] if 'target_success_level' in test_data.columns else None

print("\nFinal training features shape:", X_train.shape)
print("Final validation features shape:", X_val.shape)
print("Final test features shape:", X_test.shape)
```
